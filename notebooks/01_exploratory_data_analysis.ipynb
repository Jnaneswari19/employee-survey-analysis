# Exploratory Data Analysis

In this notebook, we perform exploratory data analysis and statistical testing on the employee survey data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import json
import os

# Load data
base_path = os.path.dirname(os.path.abspath(''))
data_path = os.path.join(base_path, 'data', 'processed', 'clean_survey_data.csv')
df = pd.read_csv(data_path)
df.head()
```

## Visualizations

```python
# Satisfaction distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['overall_satisfaction'], bins=20, kde=True)
plt.title('Distribution of Overall Satisfaction')
plt.show()
```

```python
# Department Boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(x='department', y='overall_satisfaction', data=df)
plt.title('Overall Satisfaction by Department')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(8, 6))
corr = df[['q_work_life_balance', 'q_manager_effectiveness', 'q_career_growth', 'overall_satisfaction', 'composite_engagement_score']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

## Statistical Analysis (Kruskal-Wallis H-test)

```python
# Group data by department
departments = df['department'].unique()
groups = [df[df['department'] == dept]['overall_satisfaction'] for dept in departments]

# Perform Kruskal-Wallis H-test
stat, p_value = stats.kruskal(*groups)

print(f'Kruskal-Wallis H-test statistic: {stat}')
print(f'p-value: {p_value}')

# Save results
results = {
    'kruskal_wallis_statistic': stat,
    'p_value': p_value,
    'significant': p_value < 0.05
}
output_path = os.path.join(base_path, 'notebooks', 'statistical_results.json')
with open(output_path, 'w') as f:
    json.dump(results, f, indent=4)

print(f'Saved statistical results to {output_path}')
```
